```python
def clean_data(df_clean):
    # Rename column 'brand' to 'Product company'
    df_clean = df_clean.rename(columns={'brand': 'Product company'})
    # Drop rows with missing data across all columns
    df_clean = df_clean.dropna()
    # Rename column 'breadCrumbs' to 'Product category'
    df_clean = df_clean.rename(columns={'breadCrumbs': 'Product category'})
    # Rename column 'title' to 'Product name'
    df_clean = df_clean.rename(columns={'title': 'Product name'})
    return df_clean

df_clean_1 = clean_data(df_clean.copy())
df_clean_1.head()
```
